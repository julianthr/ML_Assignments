```python
# Importing 1 examplary file
y, sr = librosa.load("C:/Users/marce/OneDrive/Dokumente/Master/CBS/2nd semester/ML/Project/Music/moodtheme_audio/00/7400.mp3")

print('y:', y, '\n')
print('Number of amplitude samples:', np.shape(y)[0], '\n')
print('Sample Rate (Hz):', sr, '\n')
```

Each value in the numpy array y represents the amplitude of the audio waveform at a specific point in time. This amplitude value is one sample of teh audio waveform.

The sample rate sr is the number of samples per second used to represent the audio waveform. It specifies how many times the amplitude of the audio waveform is measured in one second. The sample rate is typically measured in Hertz (Hz).

We can check if the audio file was loaded correctly by calculating the length of the audio in seconds.

```python
print(len(y)/sr, "s")
```

<img src="https://imgur.com/YPVEBnG.png" title="source: imgur.com" />

Together, the numpy array y and sample rate sr allow us to represent the audio waveform as a time-varying sequence of amplitude values. The amplitude corresponds to the air pressure at a given point in time. The higher the air pressure and the louder the corresponding sound.

```python
# plot raw waveform
plt.figure(figsize = (16, 6))
librosa.display.waveshow(y = y, sr = sr, axis = 's')
plt.title("Raw waveform of 7400.mp3", fontsize = 20)
plt.ylabel("Amplitude scaled to [-1,1]")
```

```python
# plot short excerpt of raw waveform
plt.figure(figsize = (16, 6))
librosa.display.waveshow(y = y[0:1000], sr = sr, axis="s");
plt.title("Raw waveform of 7400.mp3 - short excerpt", fontsize = 20)
```

### Transform into Mel Spectrogram

```python
SAMPLE_RATE = 22050
# TRACK_DURATION = 30 # measured in seconds
# SAMPLES_PER_TRACK = SAMPLE_RATE * TRACK_DURATION
hop_length = 512 # in num. of samples
n_fft = 2048 # window in num. of samples
```

```python
# transform y into melspectogram
y_melspec = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

# plot melspectogram
plt.figure(figsize = (16, 6))
librosa.display.specshow(y_melspec, sr = sr, hop_length = hop_length, x_axis = 's', y_axis = 'log')
plt.colorbar()
plt.title("Mel spectogram of 7400.mp3", fontsize = 20)
```

### Transform into Mel Frequency Cepstral Coefficients (MFCC)

```python
# transform y into MFCC
y_mfcc = librosa.feature.mfcc(y=y, sr=sr)

# plot MFCC
plt.figure(figsize = (16, 6))
librosa.display.specshow(y_mfcc, sr = sr, hop_length = hop_length, x_axis = 's', y_axis = 'log')
plt.colorbar()
plt.title("MFCC of 7400.mp3", fontsize = 20)
```

```python
# plot raw waveform split into percussive and harmonic components
y_harm, y_perc = librosa.effects.hpss(y)
plt.figure(figsize = (16, 6))
librosa.display.waveshow(y_harm, sr=sr, alpha=0.5, label='Harmonic', axis = 's')
librosa.display.waveshow(y_perc, sr=sr, color='r', alpha=0.5, label='Percussive', axis = 's')
plt.title('Raw waveform of 7400.mp3 split into harmonic and percussive components', fontsize = 20)
plt.legend()
```

The HPSS algorithm is based on the idea that harmonic sounds (such as those produced by musical instruments) have a pitch that can be estimated and modeled as a sinusoid, while percussive sounds (such as drums) are more transient and can be modeled as impulses or noise.

The algorithm decomposes the audio signal into a spectrogram, which is a time-frequency representation of the signal. Then, it applies a filter to the spectrogram that emphasizes the harmonic or percussive components, depending on which one is being separated. Finally, it converts the filtered spectrogram back into the time domain to obtain the separated components as audio signals.
